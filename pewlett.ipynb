```python
from sqlalchemy import create_engine

import sqlalchemy as sql
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


import pandas.io.sql as psql
import psycopg2
```

```python
#engine = create_engine("sqlite:///P.hachks.db")
engine = create_engine(f'postgresql+psycopg2://postgres:****@localhost:5432/P.hachks')

connection = engine.connect()

table_name = 'employees'
emp_df = pd.read_sql_table(table_name, engine)
emp_df.head()


```

```python
table_name = 'salaries'
sal_df = pd.read_sql_table(table_name, engine)
sal_df.head()
```

```python
merge_table = pd.merge(sal_df,emp_df, on = 'emp_no')

merge_table
```

```python
table_name = 'titles'
titles_df = pd.read_sql_table(table_name, engine)
titles_df.head()
```

```python
merge_table2 = pd.merge(titles_df, merge_table, on = 'emp_no')

merge_table2
```

```python
#total salary by title


salary_sum_df = merge_table2.groupby('title').sum()["salary"]

salary_sum_df
```

```python
#count how many employees per title



title_count_df = merge_table2.groupby('title').count()["emp_no"]

title_count_df
```

```python
#avg salary per title



avg_salary = salary_sum_df/title_count_df

avg_salary_title = avg_salary.round(2)
                                    
avg_salary_title
```

```python
#Summary_Table_Schools = pd.DataFrame({"Students Per School": students_per_school, "Type": school_type,"School Budget":Budget_per_school, "Budget_per_student": Per_student_budget,
                                      #"Average Math Score":average_math_df
                                      
                                      
Summary_df = pd.DataFrame({"Employees_Per_Title": title_count_df, "Total_Salary_Per_Title": salary_sum_df, "Avg_salary":avg_salary_title})                                    
                                      
Summary_df   

```

```python
new_df = Summary_df.reset_index()

new_df
```
